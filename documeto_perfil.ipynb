```python
import pandas as pd

# Definir los grupos etarios
grupos_etarios = {
    'Niño, Niña. Población de 0 a 13 años.': list(range(0, 14)),
    'Adolescente de 14 a 17 años.': list(range(14, 18)),
    'Joven de 18 a 26 años.': list(range(18, 27)),
    'Adulto entre 27 y 59 años.': list(range(27, 60)),
    'Adulto mayor. Personas igual o mayor a 60 años.': list(range(60, 101)),  # Asumimos que 100+ se incluye aquí
    'SIN DATO': []  # No hay edades específicas para este grupo
}

# Crear nuevas columnas para cada grupo etario
for grupo, edades in grupos_etarios.items():
    if edades:  # Solo agrupar si hay edades definidas
        # Crear listas para las columnas que existen
        hombres_columns = [f'Hombres_{edad}' for edad in edades if f'Hombres_{edad}' in df_poblacion.columns]
        mujeres_columns = [f'Mujeres_{edad}' for edad in edades if f'Mujeres_{edad}' in df_poblacion.columns]
        total_columns = [f'Total_{edad}' for edad in edades if f'Total_{edad}' in df_poblacion.columns]
        
        # Sumar solo las columnas que existen
        df_poblacion[f'Hombres_{grupo}'] = df_poblacion[hombres_columns].sum(axis=1)
        df_poblacion[f'Mujeres_{grupo}'] = df_poblacion[mujeres_columns].sum(axis=1)
        df_poblacion[f'Total_{grupo}'] = df_poblacion[total_columns].sum(axis=1)

# Mostrar las primeras filas del DataFrame para verificar
print(df_poblacion.head())
```

```python
from docx import Document
from docx.shared import Inches, Pt
from docx.enum.text import WD_ALIGN_PARAGRAPH
import matplotlib.pyplot as plt
import io
import geopandas as gpd
import pandas as pd
import os
import numpy as np
from itertools import product

# Configuración
ANCHO_IMAGEN = 10  # cm
MARGEN = 0.1
COLOR_MAPA = '#4e79a7'

# Contador para figuras
figura_counter = 1
tabla_counter = 1

# Filtrar datos para 2024
df_vbg_2024 = df_vbg[df_vbg['a_o_hechos'] == 2024].copy()

# Crear documento
doc = Document()

# Definir orden de grupos etarios (verifica que coincidan con tus datos)
orden_etarios = [
    'Niño, Niña. Población de 0 a 13 años.',
    'Adolescente de 14 a 17 años.',
    'Joven de 18 a 26 años.',
    'Adulto entre 27 y 59 años.',
    'Adulto mayor. Personas igual o mayor a 60 años.',
    'SIN DATO'
]

for municipio in df_vbg_2024['municipio_hecho'].unique():
    # Filtrar datos para el municipio
    df_municipio = df_vbg_2024[df_vbg_2024['municipio_hecho'] == municipio].copy()
    
    # Saltar municipios sin datos
    if df_municipio.empty:
        continue
    
    # --- Encabezado ---
    titulo = doc.add_paragraph(style='Heading 1')
    run = titulo.add_run(f"MUNICIPIO DE {municipio.upper()}")
    run.font.size = Pt(14)
    run.bold = True
    titulo.alignment = WD_ALIGN_PARAGRAPH.CENTER
    
    # --- Mapa ---
    municipio_geo = gdf_meta[gdf_meta['MPIO_CNMBR'] == municipio]
    if not municipio_geo.empty:
        fig = plt.figure(figsize=(ANCHO_IMAGEN/2.54, ANCHO_IMAGEN/2.54))
        ax = fig.add_axes([0, 0, 1, 1])
        
        bounds = municipio_geo.total_bounds
        x_margin = (bounds[2] - bounds[0]) * MARGEN
        y_margin = (bounds[3] - bounds[1]) * MARGEN
        
        ax.set_xlim(bounds[0] - x_margin, bounds[2] + x_margin)
        ax.set_ylim(bounds[1] - y_margin, bounds[3] + y_margin)
        
        municipio_geo.plot(ax=ax, color=COLOR_MAPA, edgecolor='white', linewidth=0.5)
        ax.axis('off')
        
        img_buffer = io.BytesIO()
        plt.savefig(img_buffer, format='png', dpi=300, bbox_inches='tight', pad_inches=0)
        plt.close()
        img_buffer.seek(0)
        doc.add_picture(img_buffer, width=Inches(8/2.54))

    # --- Sección de análisis poblacional ---
    # (Insertar después del mapa y antes del texto descriptivo de VBG)

    # Título de sección
    seccion_politica = doc.add_paragraph()
    run = seccion_politica.add_run("POBLACIÓN 2024")
    run.font.size = Pt(12)
    run.bold = True
    seccion_politica.alignment = WD_ALIGN_PARAGRAPH.LEFT

# Filtrar datos poblacionales para el municipio y año 2024
    df_pob_municipio = df_poblacion[
        (df_poblacion['DPMP'] == municipio) & 
        (df_poblacion['AÑO'] == 2024)
    ].copy()

    if not df_pob_municipio.empty:
        # Obtener datos básicos de población
        fila_poblacion = df_pob_municipio.iloc[0]
        
        total_poblacion = int(fila_poblacion['Total General'])
        total_mujeres = int(fila_poblacion['Total Mujeres'])
        total_hombres = int(fila_poblacion['Total Hombres'])
        
        # Definir grupos etarios y sus columnas correspondientes
        grupos_etarios_pob = {
            'Niño, Niña. Población de 0 a 13 años.': 'Total_Niño, Niña. Población de 0 a 13 años.',
            'Adolescente de 14 a 17 años.': 'Total_Adolescente de 14 a 17 años.',
            'Joven de 18 a 26 años.': 'Total_Joven de 18 a 26 años.',
            'Adulto entre 27 y 59 años.': 'Total_Adulto entre 27 y 59 años.',
            'Adulto mayor. Personas igual o mayor a 60 años.': 'Total_Adulto mayor. Personas igual o mayor a 60 años.'
        }
        
        # Calcular población por grupo etario
        poblacion_grupos = {}
        for grupo, columna in grupos_etarios_pob.items():
            if columna in fila_poblacion:
                poblacion_grupos[grupo] = int(fila_poblacion[columna])
            else:
                poblacion_grupos[grupo] = 0
        
        # Encontrar grupo etario con mayor y menor población
        grupo_mayor = max(poblacion_grupos, key=poblacion_grupos.get)
        grupo_menor = min(poblacion_grupos, key=poblacion_grupos.get)
        
        # Simplificar nombres de grupos para el texto
        nombres_simplificados = {
            'Niño, Niña. Población de 0 a 13 años.': 'niños y niñas (0-13 años)',
            'Adolescente de 14 a 17 años.': 'adolescentes (14-17 años)',
            'Joven de 18 a 26 años.': 'jóvenes (18-26 años)',
            'Adulto entre 27 y 59 años.': 'adultos (27-59 años)',
            'Adulto mayor. Personas igual o mayor a 60 años.': 'adultos mayores (60+ años)'
        }
        
        # Texto descriptivo
        texto_poblacion = f"""El municipio de {municipio.capitalize()} durante el año 2024 registraba una población total de {total_poblacion:,} habitantes, compuesta por {total_mujeres:,} mujeres ({(total_mujeres/total_poblacion)*100:.1f}%) y {total_hombres:,} hombres ({(total_hombres/total_poblacion)*100:.1f}%). Respecto a las edades de su población, evidenciamos que el grupo etario de mayor participación en el municipio es el de {nombres_simplificados[grupo_mayor]} con {poblacion_grupos[grupo_mayor]:,} habitantes, mientras que el grupo etario con menor representación es el de {nombres_simplificados[grupo_menor]} con {poblacion_grupos[grupo_menor]:,} habitantes."""
        
        doc.add_paragraph(texto_poblacion, style='BodyText')
        p.alignment = WD_ALIGN_PARAGRAPH.JUSTIFY
        
        # --- Crear pirámide poblacional ---
        # Preparar datos para la pirámide
        datos_piramide = []
        colores_grupos = ['#4e79a7', '#a0cbe8', '#f28e2b', '#e15759', '#59a14f']
        
        for i, (grupo, _) in enumerate(grupos_etarios_pob.items()):
            # Obtener datos por sexo
            col_mujeres = f'Mujeres_{grupo}'
            col_hombres = f'Hombres_{grupo}'
            
            mujeres = int(fila_poblacion.get(col_mujeres, 0)) if col_mujeres in fila_poblacion else 0
            hombres = int(fila_poblacion.get(col_hombres, 0)) if col_hombres in fila_poblacion else 0
            
            datos_piramide.append({
                'grupo': grupo,
                'mujeres': mujeres,
                'hombres': hombres,
                'color': colores_grupos[i % len(colores_grupos)]
            })
        
        # Crear la pirámide poblacional
        fig, ax = plt.subplots(figsize=(12/2.54, 10/2.54))
        plt.rcParams.update({'font.size': 8})
        
        # Preparar datos
        y_pos = range(len(datos_piramide))
        grupos_labels = [nombres_simplificados.get(d['grupo'], d['grupo']) for d in datos_piramide]
        
        
        # Crear barras horizontales
        mujeres_vals = [-d['mujeres'] for d in datos_piramide]  # Negativo para el lado izquierdo
        hombres_vals = [d['hombres'] for d in datos_piramide]
        
        # Barras
        bars_mujeres = ax.barh(y_pos, mujeres_vals, height=0.8, 
                            label='Mujeres', color='#e15759', 
                            edgecolor='white', linewidth=0.5)
        bars_hombres = ax.barh(y_pos, hombres_vals, height=0.8, 
                            label='Hombres', color='#4e79a7',
                            edgecolor='white', linewidth=0.5)
        
        # Configurar ejes
        ax.set_yticks(y_pos)
        ax.set_yticklabels(grupos_labels, fontsize=7)
        
        # Configurar eje X para mostrar valores absolutos
        max_val = max(max(abs(m) for m in mujeres_vals), max(hombres_vals))
        ax.set_xlim(-max_val * 1.1, max_val * 1.1)
        
        # Etiquetas del eje X en valores absolutos
        ticks = ax.get_xticks()
        ax.set_xticklabels([f'{abs(int(tick)):,}' for tick in ticks], fontsize=7)
        
        # Línea central
        ax.axvline(x=0, color='black', linewidth=0.8, alpha=0.8)
        
        # Estética
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.spines['bottom'].set_visible(False)
        ax.grid(axis='x', linestyle=':', alpha=0.4, linewidth=0.5)
        ax.set_axisbelow(True)
        
        # Leyenda posicionada más a la derecha
        ax.legend(frameon=False, fontsize=8, bbox_to_anchor=(1.15, 1), loc='upper left')
        
        # Etiquetas
        ax.set_xlabel('Población', fontsize=8)
        ax.text(-max_val * 0.6, -0.7, 'Mujeres', ha='center', fontsize=8, fontweight='bold')
        ax.text(max_val * 0.6, -0.7, 'Hombres', ha='center', fontsize=8, fontweight='bold')
        
        # Ajustar layout
        plt.tight_layout()
        
        # Título de la figura
        figura_title = doc.add_paragraph()
        run = figura_title.add_run(
            f"Figura {figura_counter}. Pirámide poblacional por grupo etario y sexo "
            f"del municipio de {municipio.capitalize()} (2024)"
        )
        run.font.size = Pt(11)
        run.bold = True
        figura_title.alignment = WD_ALIGN_PARAGRAPH.CENTER
        
        # Insertar gráfico
        img_buffer = io.BytesIO()
        plt.savefig(img_buffer, format='png', dpi=300, bbox_inches='tight')
        plt.close()
        img_buffer.seek(0)
        doc.add_picture(img_buffer, width=Inches(12/2.54))
        
        # Nota al pie
        nota_poblacion = doc.add_paragraph()
        run = nota_poblacion.add_run("Elaborado por el Observatorio de Mujeres y Equidad de Género con datos del DANE.")
        run.font.size = Pt(9)
        run.italic = True
        nota_poblacion.alignment = WD_ALIGN_PARAGRAPH.CENTER
        
        figura_counter += 1
        
        # --- NUEVO: Análisis de ranking poblacional y evolución histórica ---
        
        # Calcular ranking poblacional para 2024
        df_ranking_2024 = df_poblacion[df_poblacion['AÑO'] == 2024].copy()
        df_ranking_2024 = df_ranking_2024.sort_values('Total General', ascending=False).reset_index(drop=True)
        df_ranking_2024['ranking'] = df_ranking_2024.index + 1
        
        # Encontrar posición del municipio
        posicion_municipio = df_ranking_2024[df_ranking_2024['DPMP'] == municipio]['ranking'].iloc[0] if not df_ranking_2024[df_ranking_2024['DPMP'] == municipio].empty else "No disponible"
        
        # Obtener datos históricos del municipio
        df_historico_municipio = df_poblacion[df_poblacion['DPMP'] == municipio].copy()
        df_historico_municipio = df_historico_municipio.sort_values('AÑO')
        
        # Calcular crecimiento promedio anual si hay suficientes datos
        if len(df_historico_municipio) > 1:
            # Calcular crecimiento promedio para población total, mujeres y hombres
            años = df_historico_municipio['AÑO'].values
            pob_total = df_historico_municipio['Total General'].values
            pob_mujeres = df_historico_municipio['Total Mujeres'].values
            pob_hombres = df_historico_municipio['Total Hombres'].values
            
            # Crecimiento promedio anual usando regresión lineal simple
            from scipy.stats import linregress
            
            # Para población total
            slope_total, _, _, _, _ = linregress(años, pob_total)
            # Para mujeres
            slope_mujeres, _, _, _, _ = linregress(años, pob_mujeres)
            # Para hombres
            slope_hombres, _, _, _, _ = linregress(años, pob_hombres)
            
            # Convertir a enteros para mejor presentación
            crecimiento_total = int(round(slope_total))
            crecimiento_mujeres = int(round(slope_mujeres))
            crecimiento_hombres = int(round(slope_hombres))
            
            # Determinar ordinales para el ranking (29 municipios del Meta)
            ordinales = {
                1: "primer", 2: "segundo", 3: "tercer", 4: "cuarto", 5: "quinto",
                6: "sexto", 7: "séptimo", 8: "octavo", 9: "noveno", 10: "décimo",
                11: "undécimo", 12: "duodécimo", 13: "decimotercer", 14: "decimocuarto", 15: "decimoquinto",
                16: "decimosexto", 17: "decimoséptimo", 18: "decimoctavo", 19: "decimonoveno", 20: "vigésimo",
                21: "vigésimo primer", 22: "vigésimo segundo", 23: "vigésimo tercer", 24: "vigésimo cuarto", 25: "vigésimo quinto",
                26: "vigésimo sexto", 27: "vigésimo séptimo", 28: "vigésimo octavo", 29: "vigésimo noveno"
            }
            
            ordinal = ordinales.get(posicion_municipio, f"{posicion_municipio}°")
            
            # Texto del análisis de evolución poblacional
            if crecimiento_total > 0:
                tendencia_total = "crece"
            elif crecimiento_total < 0:
                tendencia_total = "decrece"
            else:
                tendencia_total = "se mantiene estable"
                
            if crecimiento_mujeres > 0:
                tendencia_mujeres = "crece"
            elif crecimiento_mujeres < 0:
                tendencia_mujeres = "decrece"
            else:
                tendencia_mujeres = "se mantiene estable"
                
            if crecimiento_hombres > 0:
                tendencia_hombres = "crece"
            elif crecimiento_hombres < 0:
                tendencia_hombres = "decrece"
            else:
                tendencia_hombres = "se mantiene estable"
            
            texto_evolucion = f"""De esta manera el municipio de {municipio.capitalize()} es el {ordinal} municipio con mayor población total en el departamento del Meta para el 2024. Si analizamos su evolución poblacional histórica, estimamos que cada año su población total {tendencia_total} en aproximadamente {abs(crecimiento_total):,} habitantes, su población de mujeres {tendencia_mujeres} en aproximadamente {abs(crecimiento_mujeres):,} habitantes, y su población de hombres {tendencia_hombres} en aproximadamente {abs(crecimiento_hombres):,} habitantes."""
            
        else:
            texto_evolucion = f"""De esta manera el municipio de {municipio.capitalize()} es el {posicion_municipio}° municipio con mayor población total en el departamento del Meta para el 2024. No se cuenta con suficientes datos históricos para estimar la evolución poblacional del municipio."""
        
        doc.add_paragraph(texto_evolucion, style='BodyText')
        p.alignment = WD_ALIGN_PARAGRAPH.JUSTIFY
        
        # --- Crear gráfica de serie de tiempo ---
        if len(df_historico_municipio) > 1:
            fig, ax = plt.subplots(figsize=(12/2.54, 8/2.54))
            plt.rcParams.update({'font.size': 8})
            
            # Datos para la gráfica
            años = df_historico_municipio['AÑO'].values
            total_general = df_historico_municipio['Total General'].values
            mujeres = df_historico_municipio['Total Mujeres'].values
            hombres = df_historico_municipio['Total Hombres'].values
            
            # Crear las líneas
            ax.plot(años, total_general, marker='o', linewidth=2, markersize=4, 
                   color='#2eab49', label='Población Total', markerfacecolor='white', markeredgewidth=1)
            ax.plot(años, mujeres, marker='s', linewidth=2, markersize=4, 
                   color='#e15759', label='Mujeres', markerfacecolor='white', markeredgewidth=1)
            ax.plot(años, hombres, marker='^', linewidth=2, markersize=4, 
                   color='#4e79a7', label='Hombres', markerfacecolor='white', markeredgewidth=1)
            
            # Configurar ejes
            ax.set_xlabel('Año', fontsize=8)
            ax.set_ylabel('Población (habitantes)', fontsize=8)
            ax.tick_params(axis='both', labelsize=7)
            
            # Formatear eje Y para mostrar números con comas
            ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{int(x):,}'))
            
            # Configurar eje X para mostrar todos los años
            ax.set_xticks(años)
            ax.set_xticklabels([str(int(año)) for año in años], rotation=45, ha='right')
            
            # Estética
            ax.spines['top'].set_visible(False)
            ax.spines['right'].set_visible(False)
            ax.grid(True, linestyle=':', alpha=0.4, linewidth=0.5)
            ax.set_axisbelow(True)
            
            # Leyenda
            ax.legend(frameon=False, fontsize=8, loc='upper left')
            
            # Ajustar layout
            plt.tight_layout()
            
            # Título de la figura
            figura_title = doc.add_paragraph()
            run = figura_title.add_run(
                f"Figura {figura_counter}. Evolución poblacional histórica "
                f"del municipio de {municipio.capitalize()} por sexo"
            )
            run.font.size = Pt(11)
            run.bold = True
            figura_title.alignment = WD_ALIGN_PARAGRAPH.CENTER
            
            # Insertar gráfico
            img_buffer = io.BytesIO()
            plt.savefig(img_buffer, format='png', dpi=300, bbox_inches='tight')
            plt.close()
            img_buffer.seek(0)
            doc.add_picture(img_buffer, width=Inches(12/2.54))
            
            # Nota al pie
            nota_evolucion = doc.add_paragraph()
            run = nota_evolucion.add_run("Elaborado por el Observatorio de Mujeres y Equidad de Género con datos del DANE.")
            run.font.size = Pt(9)
            run.italic = True
            nota_evolucion.alignment = WD_ALIGN_PARAGRAPH.CENTER
            
            figura_counter += 1
        
        # Espacio antes del análisis de VBG
        doc.add_paragraph()

    else:
        # Si no hay datos poblacionales
        doc.add_paragraph(f"No se encontraron datos poblacionales para el municipio de {municipio.capitalize()} en 2024.", style='BodyText')
        doc.add_paragraph()
    
    # --- Cálculos ---
    # Título de sección
    seccion_politica = doc.add_paragraph()
    run = seccion_politica.add_run("ANALISIS DE VIOLENCIAS BASADAS EN GÉNERO (VBG) EN 2024")
    run.font.size = Pt(12)
    run.bold = True
    seccion_politica.alignment = WD_ALIGN_PARAGRAPH.LEFT

    total = df_municipio['total_victimas'].sum()
    if total == 0:
        doc.add_paragraph(f"No se registraron víctimas en {municipio.capitalize()} durante 2024.")
        continue
    
    sexo_counts = df_municipio.groupby('sexo')['total_victimas'].sum()
    total_masculino = sexo_counts.get('MASCULINO', 0)
    total_femenino = sexo_counts.get('FEMENINO', 0)
    
    if total_masculino > total_femenino:
        sexo_mayor = 'masculino'
        tasa_sexo = f"{(total_masculino/total)*100:.1f}%"
    else:
        sexo_mayor = 'femenino'
        tasa_sexo = f"{(total_femenino/total)*100:.1f}%"
    
    # Cálculo para delitos
    delitos_data = []
    for nombre, grupo in df_municipio.groupby('grupo_delito'):
        masc = grupo[grupo['sexo'] == 'MASCULINO']['total_victimas'].sum()
        fem = grupo[grupo['sexo'] == 'FEMENINO']['total_victimas'].sum()
        delitos_data.append({
            'grupo_delito': nombre,
            'total_victimas': masc + fem,
            'total_masculino': masc,
            'total_femenino': fem
        })
    
    delitos_df = pd.DataFrame(delitos_data)
    delito_mayor = delitos_df.loc[delitos_df['total_victimas'].idxmax()]
    
    # --- Texto descriptivo ---
    texto = f"""Durante el año 2024, en el municipio de {municipio.capitalize()}, se registró un total de {total} víctimas de violencias basadas en género, {total_masculino} víctimas de sexo masculino, y {total_femenino} víctimas de sexo femenino, siendo el sexo {sexo_mayor} el que registró un mayor número de víctimas con una participación del {tasa_sexo}.

Respecto al total de víctimas por grupo de delito, encontramos que el delito con mayor participación fue {delito_mayor['grupo_delito'].lower()} con {delito_mayor['total_victimas']} víctimas en total, de las cuales {delito_mayor['total_masculino']} eran de sexo masculino y {delito_mayor['total_femenino']} de sexo femenino."""
    
    doc.add_paragraph(texto, style='BodyText')
    p.alignment = WD_ALIGN_PARAGRAPH.JUSTIFY
    
    # --- Gráfico de barras mejorado ---
    delitos_df = delitos_df.sort_values('total_victimas', ascending=False)
    
    # Configuración del gráfico
    fig, ax = plt.subplots(figsize=(10/2.54, 8/2.54))
    plt.rcParams.update({
        'font.size': 6,
        'axes.labelsize': 6,
        'xtick.labelsize': 7,
        'ytick.labelsize': 7
    })

    x = range(len(delitos_df))
    bar_width = 0.35

    # Barras
    bars_m = ax.bar([i - bar_width/2 for i in x], delitos_df['total_masculino'], 
                bar_width, label='Masculino', color='#4e79a7', edgecolor='white', linewidth=0.5)
    bars_f = ax.bar([i + bar_width/2 for i in x], delitos_df['total_femenino'], 
                bar_width, label='Femenino', color='#e15759', edgecolor='white', linewidth=0.5)

    # Añadir valores
    for bar in bars_m:
        height = bar.get_height()
        if height > 0:
            ax.text(bar.get_x() + bar.get_width()/2., height + 0.1,
                    f'{int(height)}',
                    ha='center', va='bottom', fontsize=6.5, color='black')

    for bar in bars_f:
        height = bar.get_height()
        if height > 0:
            ax.text(bar.get_x() + bar.get_width()/2., height + 0.1,
                    f'{int(height)}',
                    ha='center', va='bottom', fontsize=6.5, color='black')

    ax.set_xticks(x)
    ax.set_xticklabels(delitos_df['grupo_delito'], rotation=90, ha='center')
    ax.tick_params(axis='x', which='major', pad=2)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.grid(axis='y', linestyle=':', alpha=0.4, linewidth=0.5)
    ax.set_axisbelow(True)
    ax.legend(frameon=False, fontsize=8)
    plt.subplots_adjust(bottom=0.35, top=0.9)

    # Guardar gráfico
    img_buffer = io.BytesIO()
    plt.savefig(img_buffer, format='png', dpi=300, bbox_inches='tight')
    plt.close()
    img_buffer.seek(0)
    
    # Añadir título de la figura
    figura_title = doc.add_paragraph()
    run = figura_title.add_run(f"Figura {figura_counter}. Total de víctimas de VBG por sexo y grupo de delito del municipio de {municipio.capitalize()} (2024)")
    run.font.size = Pt(11)
    run.bold = True
    figura_title.alignment = WD_ALIGN_PARAGRAPH.CENTER
    
    # Insertar gráfico
    doc.add_picture(img_buffer, width=Inches(10/2.54))
    
    # Añadir nota
    nota = doc.add_paragraph()
    run = nota.add_run("Elaborado por el Observatorio de Mujeres y Equidad de Género con datos de la FGN.")
    run.font.size = Pt(9)
    run.italic = True
    nota.alignment = WD_ALIGN_PARAGRAPH.CENTER
    
    figura_counter += 1

    # --- Análisis de grupos de edad ---
    # Preparamos los datos garantizando todos los grupos etarios
    grupos_delito = df_municipio['grupo_delito'].unique()
    multi_idx = pd.MultiIndex.from_tuples(
        list(product(grupos_delito, orden_etarios)),
        names=['grupo_delito', 'grupo_etario']
    )

    delitos_edad = (
        df_municipio
        .groupby(['grupo_delito', 'grupo_etario'])['total_victimas']
        .sum()
        .reindex(multi_idx, fill_value=0)
        .unstack()
        .fillna(0)
        [orden_etarios]
    )

    # --- Análisis de grupos de edad integrado ---
    texto_edades = doc.add_paragraph()
    run = texto_edades.add_run("Si analizamos los grupos de edades de las víctimas, evidenciamos que para cada grupo de delito, el grupo de edad más vulnerable fue:\n\n")
    run.font.size = Pt(11)
    
    # Calculamos el grupo etario más vulnerable por cada delito
    analisis_edades = []
    for nombre_delito, grupo_delito in df_municipio.groupby('grupo_delito'):
        grupo_edad = grupo_delito.groupby('grupo_etario')['total_victimas'].sum().reset_index()
        if not grupo_edad.empty and grupo_edad['total_victimas'].sum() > 0:
            mayor_edad = grupo_edad.loc[grupo_edad['total_victimas'].idxmax()]
            analisis_edades.append({
                'delito': nombre_delito,
                'grupo_etario': mayor_edad['grupo_etario'],
                'total': mayor_edad['total_victimas']
            })
    
    # Ordenamos por el delito con más víctimas primero
    analisis_edades.sort(key=lambda x: x['total'], reverse=True)
    
    # Añadimos cada punto con formato de lista
    for analisis in analisis_edades:
        # Formateamos el texto
        delito_formateado = analisis['delito'].capitalize()
        grupo_etario_formateado = analisis['grupo_etario']
        if not grupo_etario_formateado.isupper():
            grupo_etario_formateado = grupo_etario_formateado.capitalize()
        
        texto_edades.add_run(f"• {delito_formateado}: {grupo_etario_formateado}, con un total de {analisis['total']} víctimas.\n")
        p.alignment = WD_ALIGN_PARAGRAPH.JUSTIFY
    

    
    # Configuración del gráfico
    fig, ax = plt.subplots(figsize=(10/2.54, 8/2.54))
    plt.rcParams.update({'font.size': 8})

    colores = [
        '#4e79a7', '#a0cbe8', '#f28e2b', '#e15759', '#59a14f', '#bab0ac'
    ]

    delitos_edad.plot(
        kind='bar', 
        stacked=True, 
        ax=ax, 
        color=colores, 
        edgecolor='white', 
        linewidth=0.5
    )

    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.grid(axis='y', linestyle=':', alpha=0.4)
    ax.set_axisbelow(True)
    ax.tick_params(axis='x', rotation=90)
    ax.set_xlabel('')

    leyenda_labels = [
        '0-13 años', '14-17 años', '18-26 años', 
        '27-59 años', '60+ años', 'Sin dato'
    ]
    ax.legend(
        title='Grupo etario',
        labels=leyenda_labels,
        frameon=False,
        bbox_to_anchor=(1.05, 1)
    )

    # Añadir título
    figura_title = doc.add_paragraph()
    run = figura_title.add_run(
        f"Figura {figura_counter}. Total de víctimas por grupo de edad y grupo de delito "
        f"del municipio de {municipio.capitalize()} (2024)"
    )
    run.font.size = Pt(11)
    run.bold = True
    figura_title.alignment = WD_ALIGN_PARAGRAPH.CENTER

    # Insertar gráfico
    img_buffer = io.BytesIO()
    plt.savefig(img_buffer, format='png', dpi=300, bbox_inches='tight')
    plt.close()
    img_buffer.seek(0)
    doc.add_picture(img_buffer, width=Inches(10/2.54))

    # Añadir nota
    nota2 = doc.add_paragraph()
    run = nota2.add_run("Elaborado por el Observatorio de Mujeres y Equidad de Género con datos de la FGN.")
    run.font.size = Pt(9)
    run.italic = True
    nota2.alignment = WD_ALIGN_PARAGRAPH.CENTER

    figura_counter += 1

    # --- Nueva sección: Enfoque diferencial ---
    doc.add_paragraph()  # Espacio en blanco
    
    # Texto introductorio
    texto_intro = doc.add_paragraph()
    run = texto_intro.add_run(
        "Teniendo en cuenta el enfoque diferencial para personas LGBTI, Indígenas, "
        "Afrodescendientes, y Niños, niñas y adolescentes, podemos observar que el "
        f"municipio de {municipio.capitalize()} tuvo las siguientes víctimas para "
        "cada uno de estos grupos:"
    )
    run.font.size = Pt(11)
    p.alignment = WD_ALIGN_PARAGRAPH.JUSTIFY

    # Preparamos los datos para el gráfico
    grupos_diferenciales = {
        'LGBTI': 'aplica_lgbti',
        'Indígenas': 'indigena', 
        'Afrodescendientes': 'afrodescendiente',
        'NNA': 'aplica_nna'
    }
    
    # Creamos un DataFrame con los datos agrupados
    datos_diferencial = []
    for nombre_grupo, columna in grupos_diferenciales.items():
        for delito, grupo_delito in df_municipio[df_municipio[columna] == 'SI'].groupby('grupo_delito'):
            total = grupo_delito['total_victimas'].sum()
            if total > 0:
                datos_diferencial.append({
                    'Grupo Diferencial': nombre_grupo,
                    'Delito': delito,
                    'Total': total
                })
    
    if datos_diferencial:  # Solo si hay datos
        df_diferencial = pd.DataFrame(datos_diferencial)
        
        # --- Gráfico de barras acumuladas ---
        
        titulo_grafico = doc.add_paragraph()
        run = titulo_grafico.add_run(
            f"Figura {figura_counter}. Total de víctimas con enfoque diferencial por grupo de delito "
            f"en el municipio de {municipio.capitalize()} (2024)"
        )
        run.font.size = Pt(11)
        run.bold = True
        titulo_grafico.alignment = WD_ALIGN_PARAGRAPH.CENTER
        
        # Configuración del gráfico
        fig, ax = plt.subplots(figsize=(10/2.54, 8/2.54))
        plt.rcParams.update({
            'font.size': 8,
            'axes.titlesize': 9
        })
        
        # Paleta de colores para los delitos (usa los mismos colores que en gráficos anteriores)
        colores_delitos = plt.cm.tab20.colors
        
        # Pivotear los datos para el gráfico
        pivot_df = df_diferencial.pivot_table(
            index='Grupo Diferencial', 
            columns='Delito', 
            values='Total', 
            aggfunc='sum',
            fill_value=0
        )
        
        # Ordenar por el grupo diferencial con más víctimas
        pivot_df = pivot_df.loc[pivot_df.sum(axis=1).sort_values(ascending=False).index]
        
        # Graficar
        bars = pivot_df.plot(
            kind='bar', 
            stacked=True, 
            ax=ax, 
            color=colores_delitos[:len(pivot_df.columns)],
            edgecolor='white',
            linewidth=0.5,
            width=0.7  # Ancho de las barras
        )
        
        # Ajustes estéticos
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.grid(axis='y', linestyle=':', alpha=0.4)
        ax.set_axisbelow(True)
        ax.tick_params(axis='x', rotation=90)
        ax.set_xlabel('')
        
        # Leyenda más compacta
        ax.legend(
            title='Grupo de delito',
            frameon=False,
            bbox_to_anchor=(1.05, 1),
            fontsize=8
        )
        
        # Añadir el TOTAL encima de cada barra
        for bar in bars.containers:
            # Calculamos la altura total de cada barra
            totals = pivot_df.sum(axis=1)
            
            # Añadimos las etiquetas solo al último segmento (para evitar duplicados)
            if bar == bars.containers[-1]:
                for i, total in enumerate(totals):
                    if total > 0:
                        ax.text(
                            i,  # Posición X
                            total + (0.05 * totals.max()),  # Posición Y (ligeramente arriba del máximo)
                            f'{int(total)}',  # Texto
                            ha='center', 
                            va='bottom',
                            fontsize=6,
                            fontweight='bold'
                        )
        
        # Guardar gráfico
        img_buffer = io.BytesIO()
        plt.savefig(img_buffer, format='png', dpi=300, bbox_inches='tight')
        plt.close()
        img_buffer.seek(0)
        doc.add_picture(img_buffer, width=Inches(10/2.54))
        
        # Nota al pie
        nota3 = doc.add_paragraph()
        run = nota3.add_run("Elaborado por el Observatorio de Mujeres y Equidad de Género con datos de la FGN.")
        run.font.size = Pt(9)
        run.italic = True
        nota3.alignment = WD_ALIGN_PARAGRAPH.CENTER

        figura_counter += 1
    else:
        # Mensaje si no hay datos
        doc.add_paragraph("\nNo se registraron víctimas con enfoque diferencial en este municipio.", style='BodyText')
    
    # --- Análisis histórico mejorado ---
    df_historico = df_vbg[(df_vbg['municipio_hecho'] == municipio) & 
                        (df_vbg['a_o_hechos'].between(2010, 2024))].copy()

    if not df_historico.empty:
        # Análisis por sexo
        datos_sexo = {}
        for sexo in ['FEMENINO', 'MASCULINO']:
            df_sexo = df_historico[df_historico['sexo'] == sexo]
            if not df_sexo.empty:
                serie = df_sexo.groupby('a_o_hechos')['total_victimas'].sum()
                datos_sexo[sexo] = {
                    'serie': serie,
                    'v2023': serie.get(2023, 0),
                    'v2024': serie.get(2024, 0),
                    'tendencia': None
                }
                
                # Calcular tendencia si hay más de un año
                if len(serie) > 1:
                    z = np.polyfit(serie.index.astype(int), serie.values, 1)
                    datos_sexo[sexo]['tendencia'] = "creciente" if z[0] > 0 else "decreciente"
        
        # Texto descriptivo mejorado
        if 'FEMENINO' in datos_sexo:
            datos_fem = datos_sexo['FEMENINO']
            v2023_fem = datos_fem['v2023']
            v2024_fem = datos_fem['v2024']
            tendencia_fem = datos_fem['tendencia']
            
            # Determinar si aumentó o disminuyó
            if v2024_fem > v2023_fem:
                variacion_fem = "un aumento"
                verbo_fem = "aumentando"
            elif v2024_fem < v2023_fem:
                variacion_fem = "una disminución" 
                verbo_fem = "disminuyendo"
            else:
                variacion_fem = "un comportamiento estable"
                verbo_fem = "manteniéndose"
                
            texto_historico = f"""Respecto al comportamiento histórico del total de víctimas para el municipio de {municipio.capitalize()}, evidenciamos una tendencia {tendencia_fem if tendencia_fem else 'variable'} para las víctimas mujeres. En cuanto a la variación del total de mujeres víctimas en el 2024 comparado al año anterior, observamos {variacion_fem}, {verbo_fem} de {int(v2023_fem)} víctimas en 2023 a {int(v2024_fem)} víctimas en 2024."""
            
            # Si también hay datos de hombres, añadirlos
            if 'MASCULINO' in datos_sexo:
                datos_masc = datos_sexo['MASCULINO']
                v2023_masc = datos_masc['v2023']
                v2024_masc = datos_masc['v2024']
                tendencia_masc = datos_masc['tendencia']
                
                if v2024_masc > v2023_masc:
                    variacion_masc = "un aumento"
                    verbo_masc = "aumentando"
                elif v2024_masc < v2023_masc:
                    variacion_masc = "una disminución"
                    verbo_masc = "disminuyendo" 
                else:
                    variacion_masc = "un comportamiento estable"
                    verbo_masc = "manteniéndose"
                    
                texto_historico += f""" Por su parte, las víctimas hombres muestran una tendencia {tendencia_masc if tendencia_masc else 'variable'}, con {variacion_masc} entre 2023 y 2024, {verbo_masc} de {int(v2023_masc)} a {int(v2024_masc)} víctimas."""
        
        # Añadir el texto al documento
        doc.add_paragraph(texto_historico, style='BodyText')
        p.alignment = WD_ALIGN_PARAGRAPH.JUSTIFY
        
        # Crear gráfica combinada (ambos sexos en una fila)
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15/2.54, 8/2.54))
        plt.rcParams.update({'font.size': 8})
        
        # Colores consistentes con el resto del documento
        colores = {'FEMENINO': '#e15759', 'MASCULINO': '#4e79a7'}
        nombres = {'FEMENINO': 'Mujeres', 'MASCULINO': 'Hombres'}
        
        axes = [ax1, ax2]
        sexos = ['FEMENINO', 'MASCULINO']
        
        for i, sexo in enumerate(sexos):
            if sexo in datos_sexo:
                ax = axes[i]
                serie = datos_sexo[sexo]['serie']
                color = colores[sexo]
                nombre_sexo = nombres[sexo]
                
                # Gráfico de línea con marcadores
                ax.plot(serie.index, serie.values, color=color, marker='o', 
                    linewidth=2, markersize=4, label=f'{nombre_sexo}')
                
                # Línea de tendencia si hay suficientes datos
                if len(serie) > 1:
                    z = np.polyfit(serie.index.astype(int), serie.values, 1)
                    ax.plot(serie.index, np.poly1d(z)(serie.index), '--', 
                        color=color, alpha=0.6, linewidth=1.5, label='Tendencia')
                
                # Estética
                ax.spines['top'].set_visible(False)
                ax.spines['right'].set_visible(False)
                ax.grid(axis='y', linestyle=':', alpha=0.4, linewidth=0.5)
                ax.set_axisbelow(True)
                ax.legend(frameon=False, fontsize=7)
                ax.set_ylabel('Número de víctimas', fontsize=8)
                ax.set_xlabel('Año', fontsize=8)
                
                # Configurar eje X para mostrar años enteros
                years = sorted(serie.index.tolist())
                if len(years) > 8:  # Si hay muchos años, mostrar cada 2 años
                    years_to_show = years[::2]
                    if years[-1] not in years_to_show:  # Asegurar que 2024 aparezca
                        years_to_show.append(years[-1])
                else:
                    years_to_show = years
                
                ax.set_xticks(years_to_show)
                ax.set_xticklabels([str(int(year)) for year in years_to_show])
                
                # Mejorar formato de ejes
                ax.tick_params(axis='both', which='major', labelsize=7)
                
                # Resaltar 2024 si existe
                if 2024 in serie.index:
                    ax.axvline(x=2024, color=color, alpha=0.3, linestyle=':')
                    ax.scatter([2024], [serie[2024]], color=color, s=50, zorder=5, 
                            edgecolor='white', linewidth=1)
            else:
                # Si no hay datos para este sexo, ocultar el eje
                axes[i].set_visible(False)
        
        # Ajustar espaciado
        plt.tight_layout()
        plt.subplots_adjust(wspace=0.3)
        
        # Título de la figura en el documento
        figura_title = doc.add_paragraph()
        run = figura_title.add_run(
            f"Figura {figura_counter}. Evolución histórica de víctimas de VBG por sexo "
            f"en el municipio de {municipio.capitalize()} (2010-2024)"
        )
        run.font.size = Pt(11)
        run.bold = True
        figura_title.alignment = WD_ALIGN_PARAGRAPH.CENTER
        
        # Insertar gráfico
        img_buffer = io.BytesIO()
        plt.savefig(img_buffer, format='png', dpi=300, bbox_inches='tight')
        plt.close()
        img_buffer.seek(0)
        doc.add_picture(img_buffer, width=Inches(15/2.54))
        
        # Nota al pie
        nota_historico = doc.add_paragraph()
        run = nota_historico.add_run("Elaborado por el Observatorio de Mujeres y Equidad de Género con datos de la FGN.")
        run.font.size = Pt(9)
        run.italic = True
        nota_historico.alignment = WD_ALIGN_PARAGRAPH.CENTER
        
        figura_counter += 1

    # --- NUEVO: Análisis de ranking de víctimas y tasas por cada 1.000 habitantes ---

    # Obtener datos de VBG del municipio específico para el año 2024 (ya filtrado anteriormente)
    año_analisis = 2024

    # Calcular total de víctimas del municipio en 2024 (usando df_municipio que ya está filtrado)
    if not df_municipio.empty:
        # Totales por municipio específico
        total_victimas_municipio = df_municipio['total_victimas'].sum()
        
        # Víctimas por sexo en el municipio específico
        total_mujeres_victimas = df_municipio[df_municipio['sexo'] == 'FEMENINO']['total_victimas'].sum()
        total_hombres_victimas = df_municipio[df_municipio['sexo'] == 'MASCULINO']['total_victimas'].sum()
        
        print(f"Municipio: {municipio}")
        print(f"Total víctimas: {total_victimas_municipio}")
        print(f"Víctimas mujeres: {total_mujeres_victimas}")
        print(f"Víctimas hombres: {total_hombres_victimas}")
        
        # Calcular ranking de víctimas mujeres por municipio (esto sí usa todos los municipios para comparar)
        df_ranking_vbg = df_vbg[df_vbg['a_o_hechos'] == año_analisis].copy()
        df_ranking_mujeres = df_ranking_vbg[df_ranking_vbg['sexo'] == 'FEMENINO'].copy()
        
        # Agrupar por municipio y sumar víctimas para el ranking
        ranking_mujeres_por_municipio = df_ranking_mujeres.groupby('municipio_hecho')['total_victimas'].sum().reset_index()
        ranking_mujeres_por_municipio = ranking_mujeres_por_municipio.sort_values('total_victimas', ascending=False).reset_index(drop=True)
        ranking_mujeres_por_municipio['ranking'] = ranking_mujeres_por_municipio.index + 1
        
        # Encontrar posición del municipio actual
        municipio_ranking = ranking_mujeres_por_municipio[ranking_mujeres_por_municipio['municipio_hecho'] == municipio]
        posicion_vbg = municipio_ranking['ranking'].iloc[0] if not municipio_ranking.empty else 29
        
        # Ordinales para el ranking (29 municipios del Meta)
        ordinales_vbg = {
            1: "primer", 2: "segundo", 3: "tercer", 4: "cuarto", 5: "quinto",
            6: "sexto", 7: "séptimo", 8: "octavo", 9: "noveno", 10: "décimo",
            11: "undécimo", 12: "duodécimo", 13: "decimotercer", 14: "decimocuarto", 15: "decimoquinto",
            16: "decimosexto", 17: "decimoséptimo", 18: "decimoctavo", 19: "decimonoveno", 20: "vigésimo",
            21: "vigésimo primer", 22: "vigésimo segundo", 23: "vigésimo tercer", 24: "vigésimo cuarto", 25: "vigésimo quinto",
            26: "vigésimo sexto", 27: "vigésimo séptimo", 28: "vigésimo octavo", 29: "vigésimo noveno"
        }
        
        ordinal_vbg = ordinales_vbg.get(posicion_vbg, f"{posicion_vbg}°")
        
        # Obtener datos poblacionales del municipio específico para el cálculo de tasas
        df_pob_tasa = df_poblacion[(df_poblacion['DPMP'] == municipio) & (df_poblacion['AÑO'] == año_analisis)]
        
        if not df_pob_tasa.empty:
            fila_pob_tasa = df_pob_tasa.iloc[0]
            pob_total_tasa = int(fila_pob_tasa['Total General'])
            pob_mujeres_tasa = int(fila_pob_tasa['Total Mujeres'])
            pob_hombres_tasa = int(fila_pob_tasa['Total Hombres'])
            
            print(f"Población total: {pob_total_tasa}")
            print(f"Población mujeres: {pob_mujeres_tasa}")
            print(f"Población hombres: {pob_hombres_tasa}")
            
            # Calcular tasas por cada 1.000 habitantes
            tasa_general = (total_victimas_municipio / pob_total_tasa) * 1000 if pob_total_tasa > 0 else 0
            tasa_mujeres = (total_mujeres_victimas / pob_mujeres_tasa) * 1000 if pob_mujeres_tasa > 0 else 0
            tasa_hombres = (total_hombres_victimas / pob_hombres_tasa) * 1000 if pob_hombres_tasa > 0 else 0
            
            print(f"Tasa general: {tasa_general:.2f}")
            print(f"Tasa mujeres: {tasa_mujeres:.2f}")
            print(f"Tasa hombres: {tasa_hombres:.2f}")
            
            # Texto del análisis de ranking y tasas
            texto_ranking_tasas = f"""De esta manera el municipio de {municipio.capitalize()} ocupa el {ordinal_vbg} puesto entre los municipios con mayor número de mujeres víctimas de violencias basadas en género. Sin embargo, se debe considerar el factor poblacional, dado que los municipios con mayor densidad poblacional a su vez son quienes reportan mayor número de denuncias, por ello, vamos a analizar las tasas por cada 1.000 habitantes.

Para el municipio de {municipio.capitalize()}, por cada 1.000 habitantes {tasa_general:.1f} son víctimas de violencias basadas en género, si lo analizamos por sexo, por cada 1.000 mujeres {tasa_mujeres:.1f} son víctimas de violencias basadas en género, mientras que por cada 1.000 hombres {tasa_hombres:.1f} son víctimas de estas violencias."""
            
            doc.add_paragraph(texto_ranking_tasas, style='BodyText')
            p.alignment = WD_ALIGN_PARAGRAPH.JUSTIFY

            # --- Crear gráfico de tasas por grupo de delito y sexo ---
            
            # Preparar datos para el gráfico de tasas (usando df_municipio filtrado)
            df_tasas_grafico = df_municipio[df_municipio['sexo'].isin(['MASCULINO', 'FEMENINO'])].copy()
            
            # Agrupar por grupo de delito y sexo para el municipio específico
            tasas_por_grupo = df_tasas_grafico.groupby(['grupo_delito', 'sexo'])['total_victimas'].sum().reset_index()
            
            # Calcular tasas por cada 1.000 habitantes según el sexo
            def calcular_tasa(row):
                if row['sexo'] == 'FEMENINO' and pob_mujeres_tasa > 0:
                    return (row['total_victimas'] / pob_mujeres_tasa) * 1000
                elif row['sexo'] == 'MASCULINO' and pob_hombres_tasa > 0:
                    return (row['total_victimas'] / pob_hombres_tasa) * 1000
                else:
                    return 0
            
            tasas_por_grupo['tasa'] = tasas_por_grupo.apply(calcular_tasa, axis=1)
            
            # Crear el gráfico de barras agrupadas
            fig, ax = plt.subplots(figsize=(14/2.54, 10/2.54))
            plt.rcParams.update({'font.size': 8})
            
            # Preparar datos para el gráfico
            grupos_delito = tasas_por_grupo['grupo_delito'].unique()
            x_pos = np.arange(len(grupos_delito))
            width = 0.35
            
            # Datos para cada sexo
            tasas_femenino = []
            tasas_masculino = []
            
            for grupo in grupos_delito:
                # Buscar tasa para femenino
                tasa_f_df = tasas_por_grupo[(tasas_por_grupo['grupo_delito'] == grupo) & 
                                        (tasas_por_grupo['sexo'] == 'FEMENINO')]
                tasa_f = tasa_f_df['tasa'].iloc[0] if not tasa_f_df.empty else 0
                
                # Buscar tasa para masculino
                tasa_m_df = tasas_por_grupo[(tasas_por_grupo['grupo_delito'] == grupo) & 
                                        (tasas_por_grupo['sexo'] == 'MASCULINO')]
                tasa_m = tasa_m_df['tasa'].iloc[0] if not tasa_m_df.empty else 0
                
                tasas_femenino.append(tasa_f)
                tasas_masculino.append(tasa_m)
            
            # Crear barras
            bars_femenino = ax.bar(x_pos - width/2, tasas_femenino, width, 
                                label='Femenino', color='#e15759', 
                                edgecolor='white', linewidth=0.5)
            bars_masculino = ax.bar(x_pos + width/2, tasas_masculino, width, 
                                label='Masculino', color='#4e79a7',
                                edgecolor='white', linewidth=0.5)
            
            # Configurar ejes
            ax.set_xlabel('Grupo de Delito', fontsize=8)
            ax.set_ylabel('Tasa por cada 1.000 habitantes', fontsize=8)
            ax.set_xticks(x_pos)
            
            # Ajustar etiquetas del eje X
            etiquetas_grupos = [grupo.replace(' ', '\n') if len(grupo) > 15 else grupo for grupo in grupos_delito]
            ax.set_xticklabels(etiquetas_grupos, fontsize=7, ha='center')
            
            # Formatear eje Y
            ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x:.1f}'))
            ax.tick_params(axis='y', labelsize=7)
            
            # Añadir valores en las barras
            def add_value_labels(bars):
                for bar in bars:
                    height = bar.get_height()
                    if height > 0:
                        ax.annotate(f'{height:.1f}',
                                xy=(bar.get_x() + bar.get_width() / 2, height),
                                xytext=(0, 3),  # 3 points vertical offset
                                textcoords="offset points",
                                ha='center', va='bottom',
                                fontsize=6)
            
            add_value_labels(bars_femenino)
            add_value_labels(bars_masculino)
            
            # Estética
            ax.spines['top'].set_visible(False)
            ax.spines['right'].set_visible(False)
            ax.grid(axis='y', linestyle=':', alpha=0.4, linewidth=0.5)
            ax.set_axisbelow(True)
            
            # Leyenda
            ax.legend(frameon=False, fontsize=8, loc='upper center', bbox_to_anchor=(0.5, 1.15))
            
            # Ajustar layout
            plt.tight_layout()
            
            # Título de la figura
            figura_title = doc.add_paragraph()
            run = figura_title.add_run(
                f"Figura {figura_counter}. Tasa de violencias basadas en género por cada 1.000 habitantes "
                f"según grupo de delito y sexo en {municipio.capitalize()} (2024)"
            )
            run.font.size = Pt(11)
            run.bold = True
            figura_title.alignment = WD_ALIGN_PARAGRAPH.CENTER
            
            # Insertar gráfico
            img_buffer = io.BytesIO()
            plt.savefig(img_buffer, format='png', dpi=300, bbox_inches='tight')
            plt.close()
            img_buffer.seek(0)
            doc.add_picture(img_buffer, width=Inches(10/2.54))
            
            # Nota al pie
            nota_tasas = doc.add_paragraph()
            run = nota_tasas.add_run("Elaborado por el Observatorio de Mujeres y Equidad de Género con datos de la FGN y el DANE.")
            run.font.size = Pt(9)
            run.italic = True
            nota_tasas.alignment = WD_ALIGN_PARAGRAPH.CENTER
            
            figura_counter += 1
            
        else:
            doc.add_paragraph(f"No se encontraron datos poblacionales para calcular las tasas de {municipio.capitalize()} en 2024.", style='BodyText')

    else:
        doc.add_paragraph(f"No se encontraron datos de VBG para el municipio de {municipio.capitalize()} en 2024.", style='BodyText')

    # Código a insertar después del gráfico de tasas existente
    # (agregar después de figura_counter += 1)

    # --- NUEVO: Calcular ranking de tasas de mujeres víctimas por municipio ---

    # Calcular tasas de mujeres víctimas para todos los municipios
    df_ranking_tasas = []

    # Obtener todos los municipios únicos del Meta
    municipios_meta = df_vbg[df_vbg['a_o_hechos'] == año_analisis]['municipio_hecho'].unique()

    for mpio in municipios_meta:
        # Datos de VBG del municipio
        df_mpio_vbg = df_vbg[(df_vbg['municipio_hecho'] == mpio) & 
                            (df_vbg['a_o_hechos'] == año_analisis) & 
                            (df_vbg['sexo'] == 'FEMENINO')]
        
        victimas_mujeres_mpio = df_mpio_vbg['total_victimas'].sum()
        
        # Datos poblacionales del municipio
        df_mpio_pob = df_poblacion[(df_poblacion['DPMP'] == mpio) & 
                                (df_poblacion['AÑO'] == año_analisis)]
        
        if not df_mpio_pob.empty:
            pob_mujeres_mpio = int(df_mpio_pob.iloc[0]['Total Mujeres'])
            
            # Calcular tasa por cada 1.000 mujeres
            tasa_mujeres_mpio = (victimas_mujeres_mpio / pob_mujeres_mpio) * 1000 if pob_mujeres_mpio > 0 else 0
            
            df_ranking_tasas.append({
                'municipio': mpio,
                'victimas_mujeres': victimas_mujeres_mpio,
                'poblacion_mujeres': pob_mujeres_mpio,
                'tasa_mujeres': tasa_mujeres_mpio
            })

    # Crear DataFrame y ordenar por tasa
    df_ranking_tasas = pd.DataFrame(df_ranking_tasas)
    df_ranking_tasas = df_ranking_tasas.sort_values('tasa_mujeres', ascending=False).reset_index(drop=True)
    df_ranking_tasas['ranking'] = df_ranking_tasas.index + 1

    # Encontrar posición del municipio actual
    municipio_tasa_ranking = df_ranking_tasas[df_ranking_tasas['municipio'] == municipio]
    posicion_tasa = municipio_tasa_ranking['ranking'].iloc[0] if not municipio_tasa_ranking.empty else len(df_ranking_tasas)

    # Ordinales para el ranking de tasas
    ordinales_tasas = {
        1: "primer", 2: "segundo", 3: "tercer", 4: "cuarto", 5: "quinto",
        6: "sexto", 7: "séptimo", 8: "octavo", 9: "noveno", 10: "décimo",
        11: "undécimo", 12: "duodécimo", 13: "decimotercer", 14: "decimocuarto", 15: "decimoquinto",
        16: "decimosexto", 17: "decimoséptimo", 18: "decimoctavo", 19: "decimonoveno", 20: "vigésimo",
        21: "vigésimo primer", 22: "vigésimo segundo", 23: "vigésimo tercer", 24: "vigésimo cuarto", 25: "vigésimo quinto",
        26: "vigésimo sexto", 27: "vigésimo séptimo", 28: "vigésimo octavo", 29: "vigésimo noveno"
    }

    ordinal_tasa = ordinales_tasas.get(posicion_tasa, f"{posicion_tasa}°")

    # Texto descriptivo del ranking de tasas
    texto_ranking_tasa = f"""Teniendo en cuenta lo anterior, el municipio se ubica en el puesto {posicion_tasa} entre los municipios con mayor tasa de mujeres víctimas de violencias basadas en género."""

    doc.add_paragraph(texto_ranking_tasa, style='BodyText')
    p.alignment = WD_ALIGN_PARAGRAPH.JUSTIFY

    # --- Gráfico de ranking de tasas por municipio ---

    # Preparar datos para el gráfico (mostrar todos los municipios)
    fig, ax = plt.subplots(figsize=(16/2.54, 12/2.54))
    plt.rcParams.update({'font.size': 7})

    # Colores: destacar el municipio actual
    colores = ['#ff6b35' if mpio == municipio else '#4e79a7' for mpio in df_ranking_tasas['municipio']]

    # Crear gráfico de barras horizontal
    y_pos = range(len(df_ranking_tasas))
    barras = ax.barh(y_pos, df_ranking_tasas['tasa_mujeres'], 
                    color=colores, edgecolor='white', linewidth=0.5, height=0.7)

    # Configurar ejes
    ax.set_yticks(y_pos)
    ax.set_yticklabels(df_ranking_tasas['municipio'], fontsize=6)
    ax.set_xlabel('Tasa por cada 1.000 mujeres', fontsize=8)

    # Invertir el eje Y para mostrar el mayor arriba
    ax.invert_yaxis()

    # Añadir valores en las barras
    for i, (idx, row) in enumerate(df_ranking_tasas.iterrows()):
        tasa = row['tasa_mujeres']
        if tasa > 0:
            ax.text(tasa + (max(df_ranking_tasas['tasa_mujeres']) * 0.01), i,
                    f'{tasa:.1f}', va='center', ha='left', fontsize=6)

    # Formatear eje X
    ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x:.1f}'))
    ax.tick_params(axis='x', labelsize=7)

    # Estética
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.grid(axis='x', linestyle=':', alpha=0.4, linewidth=0.5)
    ax.set_axisbelow(True)

    # Destacar el municipio actual con una línea vertical
    municipio_tasa_actual = df_ranking_tasas[df_ranking_tasas['municipio'] == municipio]['tasa_mujeres'].iloc[0] if not df_ranking_tasas[df_ranking_tasas['municipio'] == municipio].empty else 0
    if municipio_tasa_actual > 0:
        ax.axvline(x=municipio_tasa_actual, color='#ff6b35', linestyle='--', alpha=0.7, linewidth=2)

    # Leyenda personalizada
    from matplotlib.patches import Patch
    legend_elements = [
        Patch(facecolor='#ff6b35', label=f'{municipio.capitalize()}'),
        Patch(facecolor='#4e79a7', label='Otros municipios')
    ]
    ax.legend(handles=legend_elements, frameon=False, fontsize=8, loc='lower right')

    # Ajustar layout
    plt.tight_layout()

    # Título de la figura
    figura_title = doc.add_paragraph()
    run = figura_title.add_run(
        f"Figura {figura_counter}. Ranking de municipios del Meta por tasa de mujeres víctimas "
        f"de violencias basadas en género por cada 1.000 habitantes (2024)"
    )
    run.font.size = Pt(11)
    run.bold = True
    figura_title.alignment = WD_ALIGN_PARAGRAPH.CENTER

    # Insertar gráfico
    img_buffer = io.BytesIO()
    plt.savefig(img_buffer, format='png', dpi=300, bbox_inches='tight')
    plt.close()
    img_buffer.seek(0)
    doc.add_picture(img_buffer, width=Inches(16/2.54))

    # Nota al pie
    nota_ranking = doc.add_paragraph()
    run = nota_ranking.add_run("Elaborado por el Observatorio de Mujeres y Equidad de Género con datos de la FGN y el DANE.")
    run.font.size = Pt(9)
    run.italic = True
    nota_ranking.alignment = WD_ALIGN_PARAGRAPH.CENTER

    figura_counter += 1


    # --- NUEVA SECCIÓN: PARTICIPACIÓN POLÍTICA Y GOBERNANZA ---

    # Título de sección
    seccion_politica = doc.add_paragraph()
    run = seccion_politica.add_run("PARTICIPACIÓN POLÍTICA Y GOBERNANZA")
    run.font.size = Pt(12)
    run.bold = True
    seccion_politica.alignment = WD_ALIGN_PARAGRAPH.LEFT

    # Participación política
    municipio_nombre_upper = municipio.upper()

    # Obtener los datos de participación política para el municipio
    df_municipio_politica = df_politica[df_politica['MUNICIPIO'] == municipio_nombre_upper]

    if not df_municipio_politica.empty:
        # --- ANÁLISIS DE PARTICIPACIÓN EN CONSEJOS ---
        
        # Crear ranking de participación de mujeres en consejos
        df_ranking_consejo = df_politica[['MUNICIPIO', 'participación candidatas mujeres consejo']].copy()
        df_ranking_consejo = df_ranking_consejo.sort_values('participación candidatas mujeres consejo', ascending=False).reset_index(drop=True)
        df_ranking_consejo['ranking'] = range(1, len(df_ranking_consejo) + 1)
        
        # Obtener posición del municipio actual
        posicion_consejo = df_ranking_consejo[df_ranking_consejo['MUNICIPIO'] == municipio_nombre_upper]['ranking'].iloc[0]
        total_municipios = len(df_ranking_consejo)
        
        # Datos del municipio actual
        participacion_mujeres_consejo = df_municipio_politica['participación candidatas mujeres consejo'].values[0]
        candidatas_consejo = df_municipio_politica['Candidatas al consejo'].values[0]
        candidatos_consejo = df_municipio_politica['Candidatos al consejo'].values[0]
        total_candidatos_consejo = df_municipio_politica['Total candidatos al consejo'].values[0]
        
        # Calcular estadísticas comparativas
        promedio_departamental_consejo = df_politica['participación candidatas mujeres consejo'].mean()
        
        # Determinar posición relativa
        if posicion_consejo <= total_municipios * 0.33:
            posicion_relativa_consejo = "tercio superior"
        elif posicion_consejo <= total_municipios * 0.66:
            posicion_relativa_consejo = "tercio medio"
        else:
            posicion_relativa_consejo = "tercio inferior"
        
        # Comparación con el promedio
        if participacion_mujeres_consejo > promedio_departamental_consejo:
            comparacion_promedio_consejo = f"superior al promedio departamental ({promedio_departamental_consejo:.1f}%)"
        elif participacion_mujeres_consejo == promedio_departamental_consejo:
            comparacion_promedio_consejo = f"igual al promedio departamental ({promedio_departamental_consejo:.1f}%)"
        else:
            diferencia = promedio_departamental_consejo - participacion_mujeres_consejo
            comparacion_promedio_consejo = f"inferior al promedio departamental ({promedio_departamental_consejo:.1f}%), con una diferencia de {diferencia:.1f} puntos porcentuales"

        # Texto explicativo mejorado para consejos
        p = doc.add_paragraph(
            f"En el análisis de la participación política de las mujeres en las candidaturas al Consejo Municipal, "
            f"{municipio.capitalize()} registró una participación del {participacion_mujeres_consejo:.1f}% de mujeres candidatas "
            f"({candidatas_consejo} mujeres de {total_candidatos_consejo} candidatos totales). Esta cifra ubica al municipio "
            f"en la posición {posicion_consejo} de {total_municipios} municipios del departamento del Meta, "
            f"situándose en el {posicion_relativa_consejo} del ranking departamental. La participación femenina en "
            f"{municipio.capitalize()} es {comparacion_promedio_consejo}."
        )
        p.alignment = WD_ALIGN_PARAGRAPH.JUSTIFY

        # Tabla de candidatos al consejo con título mejorado
        tabla_titulo = doc.add_paragraph()
        run = tabla_titulo.add_run(f"Tabla {tabla_counter}. Participación por género en candidaturas al Consejo Municipal - {municipio.capitalize()} (2024)")
        run.font.size = Pt(11)
        run.bold = True
        tabla_titulo.alignment = WD_ALIGN_PARAGRAPH.CENTER
        tabla_counter += 1

        table_consejo = doc.add_table(rows=1, cols=4)
        table_consejo.style = 'Table Grid'
        hdr_cells = table_consejo.rows[0].cells
        hdr_cells[0].text = "Candidatas mujeres"
        hdr_cells[1].text = "Candidatos hombres"
        hdr_cells[2].text = "Total candidatos"
        hdr_cells[3].text = "% Participación mujeres"

        # Aplicar formato de encabezado
        for cell in hdr_cells:
            for paragraph in cell.paragraphs:
                for run in paragraph.runs:
                    run.font.bold = True
                    run.font.size = Pt(10)
                paragraph.alignment = WD_ALIGN_PARAGRAPH.CENTER

        row_cells = table_consejo.add_row().cells
        row_cells[0].text = str(candidatas_consejo)
        row_cells[1].text = str(candidatos_consejo)
        row_cells[2].text = str(total_candidatos_consejo)
        row_cells[3].text = f"{participacion_mujeres_consejo:.1f}%"

        # Centrar contenido de las celdas de datos
        for cell in row_cells:
            for paragraph in cell.paragraphs:
                paragraph.alignment = WD_ALIGN_PARAGRAPH.CENTER

        # Nota de fuente
        nota_consejo = doc.add_paragraph()
        run = nota_consejo.add_run("Elaborado por el Observatorio de Mujeres y Equidad de Género con datos de la Registraduría Nacional del Estado Civil.")
        run.font.size = Pt(9)
        run.italic = True
        nota_consejo.alignment = WD_ALIGN_PARAGRAPH.CENTER

        # --- GRÁFICO DE RANKING PARA CONSEJOS ---
        fig, ax = plt.subplots(figsize=(16/2.54, 12/2.54))
        plt.rcParams.update({'font.size': 7})

        # Preparar datos para el gráfico
        df_ranking_consejo['municipio_clean'] = df_ranking_consejo['MUNICIPIO'].str.title()
        
        # Colores: destacar el municipio actual
        colores = ['#ff6b35' if mpio == municipio_nombre_upper else '#4e79a7' for mpio in df_ranking_consejo['MUNICIPIO']]

        # Crear gráfico de barras horizontal
        y_pos = range(len(df_ranking_consejo))
        barras = ax.barh(y_pos, df_ranking_consejo['participación candidatas mujeres consejo'], 
                        color=colores, edgecolor='white', linewidth=0.5, height=0.7)

        # Configurar ejes
        ax.set_yticks(y_pos)
        ax.set_yticklabels(df_ranking_consejo['municipio_clean'], fontsize=6)
        ax.set_xlabel('Porcentaje de participación de mujeres (%)', fontsize=8)

        # Invertir el eje Y para mostrar el mayor arriba
        ax.invert_yaxis()

        # Añadir valores en las barras
        for i, (idx, row) in enumerate(df_ranking_consejo.iterrows()):
            porcentaje = row['participación candidatas mujeres consejo']
            if porcentaje > 0:
                ax.text(porcentaje + (max(df_ranking_consejo['participación candidatas mujeres consejo']) * 0.01), i,
                        f'{porcentaje:.1f}%', va='center', ha='left', fontsize=6)

        # Línea del promedio departamental
        ax.axvline(x=promedio_departamental_consejo, color='#e74c3c', linestyle=':', alpha=0.8, linewidth=2)

        # Formatear eje X
        ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x:.0f}%'))
        ax.tick_params(axis='x', labelsize=7)

        # Estética
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.grid(axis='x', linestyle=':', alpha=0.4, linewidth=0.5)
        ax.set_axisbelow(True)

        # Leyenda personalizada
        from matplotlib.patches import Patch
        legend_elements = [
            Patch(facecolor='#ff6b35', label=f'{municipio.capitalize()}'),
            Patch(facecolor='#4e79a7', label='Otros municipios'),
            Patch(facecolor='#e74c3c', label='Promedio departamental')
        ]
        # Ubicar la leyenda arriba, fuera del gráfico
        ax.legend(
            handles=legend_elements,
            frameon=False,
            fontsize=7,
            loc='lower center',
            bbox_to_anchor=(0.5, 1.02),
            ncol=3  # para que la leyenda esté en una sola fila
        )

        plt.tight_layout()

        # Título de la figura
        figura_title = doc.add_paragraph()
        run = figura_title.add_run(
            f"Figura {figura_counter}. Ranking de participación de mujeres en candidaturas al Consejo Municipal por municipio del Meta (2024)"
        )
        run.font.size = Pt(11)
        run.bold = True
        figura_title.alignment = WD_ALIGN_PARAGRAPH.CENTER
        figura_counter += 1

        # Insertar gráfico
        img_buffer = io.BytesIO()
        plt.savefig(img_buffer, format='png', dpi=300, bbox_inches='tight')
        plt.close()
        img_buffer.seek(0)
        doc.add_picture(img_buffer, width=Inches(16/2.54))

        # Nota al pie del gráfico
        nota_ranking_consejo = doc.add_paragraph()
        run = nota_ranking_consejo.add_run("Elaborado por el Observatorio de Mujeres y Equidad de Género con datos de la Registraduría Nacional del Estado Civil.")
        run.font.size = Pt(9)
        run.italic = True
        nota_ranking_consejo.alignment = WD_ALIGN_PARAGRAPH.CENTER

        # --- ANÁLISIS DE PARTICIPACIÓN EN ALCALDÍAS ---
        
        # Crear ranking de participación de mujeres en alcaldías
        df_ranking_alcaldia = df_politica[['MUNICIPIO', 'participación candidatas mujeres alcaldía']].copy()
        df_ranking_alcaldia = df_ranking_alcaldia.sort_values('participación candidatas mujeres alcaldía', ascending=False).reset_index(drop=True)
        df_ranking_alcaldia['ranking'] = range(1, len(df_ranking_alcaldia) + 1)
        
        # Obtener posición del municipio actual
        posicion_alcaldia = df_ranking_alcaldia[df_ranking_alcaldia['MUNICIPIO'] == municipio_nombre_upper]['ranking'].iloc[0]
        
        # Datos del municipio actual
        participacion_mujeres_alcaldia = df_municipio_politica['participación candidatas mujeres alcaldía'].values[0]
        candidatas_alcaldia = df_municipio_politica['Candidatas alcaldía'].values[0]
        candidatos_alcaldia = df_municipio_politica['Candidatos alcaldía'].values[0]
        total_candidatos_alcaldia = df_municipio_politica['Total candidatos alcaldía'].values[0]
        
        # Calcular estadísticas comparativas
        promedio_departamental_alcaldia = df_politica['participación candidatas mujeres alcaldía'].mean()
        
        # Determinar posición relativa
        if posicion_alcaldia <= total_municipios * 0.33:
            posicion_relativa_alcaldia = "tercio superior"
        elif posicion_alcaldia <= total_municipios * 0.66:
            posicion_relativa_alcaldia = "tercio medio"
        else:
            posicion_relativa_alcaldia = "tercio inferior"
        
        # Comparación con el promedio
        if participacion_mujeres_alcaldia > promedio_departamental_alcaldia:
            comparacion_promedio_alcaldia = f"superior al promedio departamental ({promedio_departamental_alcaldia:.1f}%)"
        elif participacion_mujeres_alcaldia == promedio_departamental_alcaldia:
            comparacion_promedio_alcaldia = f"igual al promedio departamental ({promedio_departamental_alcaldia:.1f}%)"
        else:
            diferencia = promedio_departamental_alcaldia - participacion_mujeres_alcaldia
            comparacion_promedio_alcaldia = f"inferior al promedio departamental ({promedio_departamental_alcaldia:.1f}%), con una diferencia de {diferencia:.1f} puntos porcentuales"

        # Texto explicativo mejorado para alcaldías
        p = doc.add_paragraph(
            f"En cuanto a las candidaturas a la Alcaldía Municipal, {municipio.capitalize()} presentó una participación "
            f"del {participacion_mujeres_alcaldia:.1f}% de mujeres candidatas ({candidatas_alcaldia} mujeres de "
            f"{total_candidatos_alcaldia} candidatos totales). Esta participación posiciona al municipio en el lugar "
            f"{posicion_alcaldia} de {total_municipios} municipios del departamento, ubicándose en el {posicion_relativa_alcaldia} "
            f"del ranking departamental. La participación femenina en las candidaturas a la alcaldía de "
            f"{municipio.capitalize()} es {comparacion_promedio_alcaldia}."
        )
        p.alignment = WD_ALIGN_PARAGRAPH.JUSTIFY

        # Tabla de candidatos a la Alcaldía con título mejorado
        tabla_titulo = doc.add_paragraph()
        run = tabla_titulo.add_run(f"Tabla {tabla_counter}. Participación por género en candidaturas a la Alcaldía Municipal - {municipio.capitalize()} (2024)")
        run.font.size = Pt(11)
        run.bold = True
        tabla_titulo.alignment = WD_ALIGN_PARAGRAPH.CENTER
        tabla_counter += 1

        table_alcaldia = doc.add_table(rows=1, cols=4)
        table_alcaldia.style = 'Table Grid'
        hdr_cells = table_alcaldia.rows[0].cells
        hdr_cells[0].text = "Candidatas mujeres"
        hdr_cells[1].text = "Candidatos hombres"
        hdr_cells[2].text = "Total candidatos"
        hdr_cells[3].text = "% Participación mujeres"

        # Aplicar formato de encabezado
        for cell in hdr_cells:
            for paragraph in cell.paragraphs:
                for run in paragraph.runs:
                    run.font.bold = True
                    run.font.size = Pt(10)
                paragraph.alignment = WD_ALIGN_PARAGRAPH.CENTER

        row_cells = table_alcaldia.add_row().cells
        row_cells[0].text = str(candidatas_alcaldia)
        row_cells[1].text = str(candidatos_alcaldia)
        row_cells[2].text = str(total_candidatos_alcaldia)
        row_cells[3].text = f"{participacion_mujeres_alcaldia:.1f}%"

        # Centrar contenido de las celdas de datos
        for cell in row_cells:
            for paragraph in cell.paragraphs:
                paragraph.alignment = WD_ALIGN_PARAGRAPH.CENTER

        # Nota de fuente
        nota_alcaldia = doc.add_paragraph()
        run = nota_alcaldia.add_run("Elaborado por el Observatorio de Mujeres y Equidad de Género con datos de la Registraduría Nacional del Estado Civil.")
        run.font.size = Pt(9)
        run.italic = True
        nota_alcaldia.alignment = WD_ALIGN_PARAGRAPH.CENTER

        # --- GRÁFICO DE RANKING PARA ALCALDÍAS ---
        fig, ax = plt.subplots(figsize=(16/2.54, 12/2.54))
        plt.rcParams.update({'font.size': 7})

        # Preparar datos para el gráfico
        df_ranking_alcaldia['municipio_clean'] = df_ranking_alcaldia['MUNICIPIO'].str.title()
        
        # Colores: destacar el municipio actual
        colores = ['#ff6b35' if mpio == municipio_nombre_upper else '#4e79a7' for mpio in df_ranking_alcaldia['MUNICIPIO']]

        # Crear gráfico de barras horizontal
        y_pos = range(len(df_ranking_alcaldia))
        barras = ax.barh(y_pos, df_ranking_alcaldia['participación candidatas mujeres alcaldía'], 
                        color=colores, edgecolor='white', linewidth=0.5, height=0.7)

        # Configurar ejes
        ax.set_yticks(y_pos)
        ax.set_yticklabels(df_ranking_alcaldia['municipio_clean'], fontsize=6)
        ax.set_xlabel('Porcentaje de participación de mujeres (%)', fontsize=8)

        # Invertir el eje Y para mostrar el mayor arriba
        ax.invert_yaxis()

        # Añadir valores en las barras
        for i, (idx, row) in enumerate(df_ranking_alcaldia.iterrows()):
            porcentaje = row['participación candidatas mujeres alcaldía']
            if porcentaje > 0:
                ax.text(porcentaje + (max(df_ranking_alcaldia['participación candidatas mujeres alcaldía']) * 0.01), i,
                        f'{porcentaje:.1f}%', va='center', ha='left', fontsize=6)

        # Línea del promedio departamental
        ax.axvline(x=promedio_departamental_alcaldia, color='#e74c3c', linestyle=':', alpha=0.8, linewidth=2)


        # Formatear eje X
        ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x:.0f}%'))
        ax.tick_params(axis='x', labelsize=7)

        # Estética
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.grid(axis='x', linestyle=':', alpha=0.4, linewidth=0.5)
        ax.set_axisbelow(True)

        # Leyenda personalizada
        legend_elements = [
            Patch(facecolor='#ff6b35', label=f'{municipio.capitalize()}'),
            Patch(facecolor='#4e79a7', label='Otros municipios'),
            Patch(facecolor='#e74c3c', label='Promedio departamental')
        ]
        # Ubicar la leyenda arriba, fuera del gráfico
        ax.legend(
            handles=legend_elements,
            frameon=False,
            fontsize=7,
            loc='lower center',
            bbox_to_anchor=(0.5, 1.02),
            ncol=3  # para que la leyenda esté en una sola fila
        )
        plt.tight_layout()

        # Título de la figura
        figura_title = doc.add_paragraph()
        run = figura_title.add_run(
            f"Figura {figura_counter}. Ranking de participación de mujeres en candidaturas a la Alcaldía Municipal por municipio del Meta (2024)"
        )
        run.font.size = Pt(11)
        run.bold = True
        figura_title.alignment = WD_ALIGN_PARAGRAPH.CENTER
        figura_counter += 1

        # Insertar gráfico
        img_buffer = io.BytesIO()
        plt.savefig(img_buffer, format='png', dpi=300, bbox_inches='tight')
        plt.close()
        img_buffer.seek(0)
        doc.add_picture(img_buffer, width=Inches(16/2.54))

        # Nota al pie del gráfico
        nota_ranking_alcaldia = doc.add_paragraph()
        run = nota_ranking_alcaldia.add_run("Elaborado por el Observatorio de Mujeres y Equidad de Género con datos de la Registraduría Nacional del Estado Civil.")
        run.font.size = Pt(9)
        run.italic = True
        nota_ranking_alcaldia.alignment = WD_ALIGN_PARAGRAPH.CENTER

    else:
        no_datos_politica = doc.add_paragraph("No se encontraron datos de participación política para este municipio en 2024.")
        no_datos_politica.alignment = WD_ALIGN_PARAGRAPH.JUSTIFY

    # --- SECCIÓN DE GOBERNANZA ---

    # Filtrar los funcionarios del municipio actual
    df_municipio_funcionarios = df_funcionarios[df_funcionarios['municipioubicacionentidad'] == municipio_nombre_upper.upper()]

    # Verificar si hay datos de funcionarios para el municipio
    if not df_municipio_funcionarios.empty:
        # Calcular estadísticas básicas
        total_funcionarios = len(df_municipio_funcionarios)
        total_mujeres_funcionarios = len(df_municipio_funcionarios[df_municipio_funcionarios['sexo'] == 'FEMENINO'])
        total_hombres_funcionarios = len(df_municipio_funcionarios[df_funcionarios['sexo'] == 'MASCULINO'])
        porcentaje_mujeres_funcionarios = round((total_mujeres_funcionarios / total_funcionarios) * 100, 2) if total_funcionarios > 0 else 0

        # Crear ranking de participación de mujeres funcionarias por municipio
        funcionarios_por_municipio = df_funcionarios.groupby('municipioubicacionentidad').agg({
            'sexo': ['count', lambda x: (x == 'FEMENINO').sum()]
        }).round(2)
        
        funcionarios_por_municipio.columns = ['total_funcionarios', 'mujeres_funcionarias']
        funcionarios_por_municipio['porcentaje_mujeres'] = (funcionarios_por_municipio['mujeres_funcionarias'] / 
                                                        funcionarios_por_municipio['total_funcionarios'] * 100).round(2)
        funcionarios_por_municipio = funcionarios_por_municipio.reset_index()
        funcionarios_por_municipio = funcionarios_por_municipio.sort_values('porcentaje_mujeres', ascending=False).reset_index(drop=True)
        funcionarios_por_municipio['ranking'] = range(1, len(funcionarios_por_municipio) + 1)

        # Obtener posición del municipio actual
        posicion_funcionarios = funcionarios_por_municipio[funcionarios_por_municipio['municipioubicacionentidad'] == municipio_nombre_upper]['ranking'].iloc[0]
        total_municipios_funcionarios = len(funcionarios_por_municipio)
        promedio_departamental_funcionarios = funcionarios_por_municipio['porcentaje_mujeres'].mean()

        # Determinar posición relativa
        if posicion_funcionarios <= total_municipios_funcionarios * 0.33:
            posicion_relativa_funcionarios = "tercio superior"
        elif posicion_funcionarios <= total_municipios_funcionarios * 0.66:
            posicion_relativa_funcionarios = "tercio medio"
        else:
            posicion_relativa_funcionarios = "tercio inferior"

        # Comparación con el promedio
        if porcentaje_mujeres_funcionarios > promedio_departamental_funcionarios:
            comparacion_promedio_funcionarios = f"superior al promedio departamental ({promedio_departamental_funcionarios:.1f}%)"
        elif porcentaje_mujeres_funcionarios == promedio_departamental_funcionarios:
            comparacion_promedio_funcionarios = f"igual al promedio departamental ({promedio_departamental_funcionarios:.1f}%)"
        else:
            diferencia = promedio_departamental_funcionarios - porcentaje_mujeres_funcionarios
            comparacion_promedio_funcionarios = f"inferior al promedio departamental ({promedio_departamental_funcionarios:.1f}%), con una diferencia de {diferencia:.1f} puntos porcentuales"

        # Texto explicativo mejorado para gobernanza
        p = doc.add_paragraph(
            f"En el análisis de gobernanza y equidad de género en la administración pública municipal, basado en datos "
            f"del Departamento Administrativo de la Función Pública a través del Sistema de Información y Gestión del "
            f"Empleo Público (SIGEP), se observa que {municipio.capitalize()} cuenta con {total_funcionarios} "
            f"servidores públicos con vinculación activa, de los cuales {total_mujeres_funcionarios} son mujeres "
            f"({porcentaje_mujeres_funcionarios}%) y {total_hombres_funcionarios} son hombres "
            f"({100-porcentaje_mujeres_funcionarios:.1f}%). Esta distribución ubica al municipio en la posición "
            f"{posicion_funcionarios} de {total_municipios_funcionarios} municipios del departamento del Meta, "
            f"situándose en el {posicion_relativa_funcionarios} del ranking departamental. La participación femenina "
            f"en la planta de funcionarios públicos de {municipio.capitalize()} es {comparacion_promedio_funcionarios}."
        )
        p.alignment = WD_ALIGN_PARAGRAPH.JUSTIFY

        # Crear tabla de funcionarios públicos
        tabla_titulo = doc.add_paragraph()
        run = tabla_titulo.add_run(f"Tabla {tabla_counter}. Distribución por género de funcionarios públicos - {municipio.capitalize()} (2024)")
        run.font.size = Pt(11)
        run.bold = True
        tabla_titulo.alignment = WD_ALIGN_PARAGRAPH.CENTER
        tabla_counter += 1

        table_funcionarios = doc.add_table(rows=1, cols=4)
        table_funcionarios.style = 'Table Grid'
        hdr_cells = table_funcionarios.rows[0].cells
        hdr_cells[0].text = "Funcionarias mujeres"
        hdr_cells[1].text = "Funcionarios hombres"
        hdr_cells[2].text = "Total funcionarios"
        hdr_cells[3].text = "% Participación mujeres"

        # Aplicar formato de encabezado
        for cell in hdr_cells:
            for paragraph in cell.paragraphs:
                for run in paragraph.runs:
                    run.font.bold = True
                    run.font.size = Pt(10)
                paragraph.alignment = WD_ALIGN_PARAGRAPH.CENTER

        row_cells = table_funcionarios.add_row().cells
        row_cells[0].text = str(total_mujeres_funcionarios)
        row_cells[1].text = str(total_hombres_funcionarios)
        row_cells[2].text = str(total_funcionarios)
        row_cells[3].text = f"{porcentaje_mujeres_funcionarios:.1f}%"

        # Centrar contenido de las celdas de datos
        for cell in row_cells:
            for paragraph in cell.paragraphs:
                paragraph.alignment = WD_ALIGN_PARAGRAPH.CENTER

        # Nota de fuente para la tabla
        nota_funcionarios_tabla = doc.add_paragraph()
        run = nota_funcionarios_tabla.add_run("Elaborado por el Observatorio de Mujeres y Equidad de Género con datos del Departamento Administrativo de la Función Pública.")
        run.font.size = Pt(9)
        run.italic = True
        nota_funcionarios_tabla.alignment = WD_ALIGN_PARAGRAPH.CENTER

        # --- GRÁFICO DE RANKING PARA FUNCIONARIOS ---
        fig, ax = plt.subplots(figsize=(16/2.54, 12/2.54))
        plt.rcParams.update({'font.size': 7})

        # Preparar datos para el gráfico
        funcionarios_por_municipio['municipio_clean'] = funcionarios_por_municipio['municipioubicacionentidad'].str.title()
        
        # Colores: destacar el municipio actual
        colores = ['#ff6b35' if mpio == municipio_nombre_upper else '#4e79a7' for mpio in funcionarios_por_municipio['municipioubicacionentidad']]

        # Crear gráfico de barras horizontal
        y_pos = range(len(funcionarios_por_municipio))
        barras = ax.barh(y_pos, funcionarios_por_municipio['porcentaje_mujeres'], 
                        color=colores, edgecolor='white', linewidth=0.5, height=0.7)

        # Configurar ejes
        ax.set_yticks(y_pos)
        ax.set_yticklabels(funcionarios_por_municipio['municipio_clean'], fontsize=6)
        ax.set_xlabel('Porcentaje de funcionarias mujeres (%)', fontsize=8)

        # Invertir el eje Y para mostrar el mayor arriba
        ax.invert_yaxis()

        # Añadir valores en las barras
        for i, (idx, row) in enumerate(funcionarios_por_municipio.iterrows()):
            porcentaje = row['porcentaje_mujeres']
            if porcentaje > 0:
                ax.text(porcentaje + (max(funcionarios_por_municipio['porcentaje_mujeres']) * 0.01), i,
                        f'{porcentaje:.1f}%', va='center', ha='left', fontsize=6)

        # Línea del promedio departamental
        ax.axvline(x=promedio_departamental_funcionarios, color='#e74c3c', linestyle=':', alpha=0.8, linewidth=2)


        # Formatear eje X
        ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x:.0f}%'))
        ax.tick_params(axis='x', labelsize=7)

        # Estética
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.grid(axis='x', linestyle=':', alpha=0.4, linewidth=0.5)
        ax.set_axisbelow(True)

        # Leyenda personalizada
        legend_elements = [
            Patch(facecolor='#ff6b35', label=f'{municipio.capitalize()}'),
            Patch(facecolor='#4e79a7', label='Otros municipios'),
            Patch(facecolor='#e74c3c', label='Promedio departamental')
        ]
        # Ubicar la leyenda arriba, fuera del gráfico
        ax.legend(
            handles=legend_elements,
            frameon=False,
            fontsize=7,
            loc='lower center',
            bbox_to_anchor=(0.5, 1.02),
            ncol=3  # para que la leyenda esté en una sola fila
        )

        # Ajustar el layout para evitar recortes
        plt.tight_layout()

        # Título de la figura
        figura_title = doc.add_paragraph()
        run = figura_title.add_run(
            f"Figura {figura_counter}. Ranking de participación de mujeres en la función pública por municipio del Meta (2024)"
        )
        run.font.size = Pt(11)
        run.bold = True
        figura_title.alignment = WD_ALIGN_PARAGRAPH.CENTER
        figura_counter += 1

        # Insertar gráfico
        img_buffer = io.BytesIO()
        plt.savefig(img_buffer, format='png', dpi=300, bbox_inches='tight')
        plt.close()
        img_buffer.seek(0)
        doc.add_picture(img_buffer, width=Inches(16/2.54))

        # Nota al pie del gráfico
        nota_ranking_funcionarios = doc.add_paragraph()
        run = nota_ranking_funcionarios.add_run("Elaborado por el Observatorio de Mujeres y Equidad de Género con datos del Departamento Administrativo de la Función Pública.")
        run.font.size = Pt(9)
        run.italic = True
        nota_ranking_funcionarios.alignment = WD_ALIGN_PARAGRAPH.CENTER

    doc.add_page_break()  # Salto de página para separar municipios

doc.save('Informe_VBG_Final_Completo.docx')
print("✅ Documento generado exitosamente")
```
